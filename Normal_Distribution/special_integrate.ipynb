```python
def my_Gaussian_int(x1,x2,A,mu,sigma):
    # item1=erf((x2-mu)/(scipy.sqrt(2)*sigma))
    # item2=erf((x1-mu)/(scipy.sqrt(2)*sigma))
    # return (item1-item2)*sigma*A*scipy.sqrt(scipy.pi/2)
    item1=erf((x2-mu)/(np.sqrt(2)*sigma))
    item2=erf((x1-mu)/(np.sqrt(2)*sigma))
    return (item1-item2)*sigma*A*np.sqrt(np.pi/2)
```
